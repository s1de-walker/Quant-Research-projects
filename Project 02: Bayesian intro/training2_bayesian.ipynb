```python
import pandas as pd
```

```python
df = pd.DataFrame({
    'Signal': [1,0,0,1,1,0,1,0],
    'Returns': [0.06, 0.02,-0.03,0,-0.01,0.02,0.02,0.01],
})
```

```python
df
```

### **P(Returns > 0 | Signal =1)**

```python
condition = df["Signal"] == 1
df[condition]
```

```python
shrinked_df = df[condition]

(shrinked_df["Returns"]>0).mean()
```

### **P(Signal = 1 | Returns > 0)**

```python
# Byesian
condition2 = df["Returns"] > 0

shrinked_df2 = df[condition2]
shrinked_df2
```

```python
shrinked_df2["Signal"] == 1
(shrinked_df2["Signal"] == 1).mean()
```
