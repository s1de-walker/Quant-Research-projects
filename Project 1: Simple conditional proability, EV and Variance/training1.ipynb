```python
import pandas as pd
```

```python
df = pd.DataFrame({
    'Signal': [1,0,0,1,1,0],
    'Returns': [0.06, 0.02,-0.01,0,-0.01,0.02],
})
```

```python
df
```

### **P(Returns > 0 | Signal =1)**

```python
# condition that shrinks universe: signal = 1
signal_condition = df['Signal'] == 1
df[signal_condition]

# lets store this required universe in a separate df
signal_df = df[signal_condition]
signal_df
```

```python
# probability where returns is > 0 is sum/total
signal_df["Returns"]>0

# True = 1 nad False = 0 | sum of True / total can be calculated by mean | mean = (1 + 0 + 0) / 3 which is also the probability required.
(signal_df["Returns"]>0).mean()
```

### **E(Returns | Signal = 1)**

```python
# get the universe
signal = df["Signal"] == 1
signal_df = df[signal]
signal_df
```

Expected value is the probability weighted sum  
here the events are assumed to be equal likely  
so we can calculate average

r1 * 1/3 + r2 * 1/3 + r3 *1/3

same thing

```python
signal_df["Returns"].mean()
```

### **Variance of the above**

```python
signal_df["Returns"].var()
```

### **Standard deviation**

```python
signal_df["Returns"].std()
```
