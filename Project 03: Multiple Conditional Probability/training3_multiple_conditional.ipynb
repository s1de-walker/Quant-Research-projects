```python
import pandas as pd
```

```python
df = pd.DataFrame({
    'Signal': [1, 1, 0, 1, 0, 1],
    'Return': [0.04, -0.01, 0.02, 0.03, -0.05, 0.01],
    'VolumeSpike': [1, 0, 1, 1, 0, 1]
})

df
```

### **P(Returns > 0 | Signal =1, Vol spike = 1)**

```python
shrink_condition = (df["Signal"] == 1) & (df["VolumeSpike"] == 1)
```

```python
shrinked_df = df[shrink_condition]
shrinked_df
```

```python
(shrinked_df["Return"] > 0).mean()
```
